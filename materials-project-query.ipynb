```python
def get_elements_to_exclude(to_include: set):
    """Args:
        to_include (set of int): Atomic numbers of elements to include
    Returns:
        (list of str): Symbols of all elements to exclude, as taken
            by the Materials Project API."""
    ret = []
    for i in range(1, 112):
        if i not in to_include:
            ret.append(str(table.get_el_sp(i)))
    return ret
```

```python
include = {8, 12, 27, 28, 29, 30} # O, Mg, Co, Ni, Cu, Zn
exclude = get_elements_to_exclude(include)

with MPRester(key) as mpr:
    docs = mpr.summary.search(
        elements = ['O'],
        exclude_elements = exclude,
        fields = ['material_id', 'formula_pretty']
    )
    mpid_formulas = {
        doc.material_id: doc.formula_pretty for doc in docs
    }
```
